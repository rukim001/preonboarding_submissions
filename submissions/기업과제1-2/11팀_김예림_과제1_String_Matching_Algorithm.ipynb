```python
class make_inputs:
    
    def int_check(self):
        while True:
            try:
                n_ = int(input('입력 받을 데이터의 수를 숫자로 입력해주세요.: '))
                return n_
                break
            except:
                print('입력이 정확하지 않습니다.')
                continue
    
    def int_size_check(self):

        while True:

            n_ = self.int_check()

            if n_ < 1:
                print('너무 작습니다.')
                continue
            elif n_ > 1001:
                print('너무 큽니다.')
                continue
            else:
                print(f'입력 받을 데이터의 수 : {n_}')
                break
        
        return n_
    
    def str_size_check(self, n_, tp):
        
        if tp == 'set':
            R = set()
        else:
            R = []

        for i in range(n_):
            while True:
                elm = input('문자열을 입력해주세요.: ')

                if len(elm) > 101:
                    print('문자열의 길이가 너무 깁니다.')
                    continue
                elif len(elm) < 1:
                    print('문자열의 길이가 너무 짧습니다.')
                    continue
                else:
                    if tp == 'set':
                        R.add(elm)
                    else:
                        R.append(elm)
                    break

        return R

    def make_input(self):
    
        S = set()
        L = []
        print('************문자열 집합 생성************')
        N = self.int_size_check()
        tp = 'set'
        S = self.str_size_check(N, tp)

        print('')
        print(S)

        print('')
        print('************매칭 대상 문자열 리스트 생성************')
        Q = self.int_size_check()
        tp = 'list'
        L = self.str_size_check(Q, tp)

        print('')
        print(L)

        return S, L
```

```python
dataset = make_inputs().make_input()
```

```python
class matching_alg():

    def get_pi(self, p):
        m = len(p)
        idx = 0
        pi = [0 for i in range(m)]

        for i in range(1, m):
            while idx > 0 and p[i] != p[idx]:
                idx = pi[idx - 1]

            if p[i] == p[idx]:
                idx += 1
                pi[i] = idx
        return pi

    def kmp_r(self, s, p):
        n = len(s)
        m = len(p)
        pi = self.get_pi(p)
        idx = 0
        result = []

        for i in range(0, n):
            while idx > 0 and s[i] != p[idx]:
                idx = pi[idx - 1]
            if s[i] == p[idx]:
                if idx == m - 1:
                    result.append([1, i - idx])
                    return result
                else:
                    idx += 1
        result.append([0, -1])
        return result
    
    def kmp_nr(self, s, p):
        n = len(s)
        m = len(p)
        pi = self.get_pi(p)
        idx = 0
        result = []

        for i in range(0, n):
            while idx > 0 and s[i] != p[idx]:
                idx = pi[idx - 1]
            if s[i] == p[idx]:
                if idx == m - 1:
                    print(f'#pos = {i - idx}, pattern = "{p}"')
                else:
                    idx += 1

    def match_str(self, dataset, tp):
        input_set = dataset[0]
        input_list = dataset[1]
        
        for input_str in input_list:
            print(f'start matching in "{input_str}"')
            for elm in input_set:
                if tp == 1:
                    result = self.kmp_r(input_str, elm)
                    if result[0][0] == 1:
                        print(f'Is "{elm}" matched in "{input_str}"? : YES')
                    else:
                        print(f'Is "{elm}" matched in "{input_str}"? : NO')
                if tp == 2:
                    result = self.kmp_r(input_str, elm)
                    if result[0][1] > -1:
                        print(f'#pos = {result[0][1]}, pattern = "{elm}"')
                if tp == 3:
                    self.kmp_nr(input_str, elm)
    
    def fit_all(self, dataset):
        self.match_str(dataset, 1)
        print('')
        self.match_str(dataset, 2)
        print('')
        self.match_str(dataset, 3)
    
```

```python
matching_sys = matching_alg()
matching_sys.fit_all(dataset)
```
